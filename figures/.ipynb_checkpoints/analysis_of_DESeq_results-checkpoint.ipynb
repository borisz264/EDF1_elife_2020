### 20200409: make plots of hits from DEseq analysis

First I need to assemble the data from salmon (TPMs) and DEseq2 (fold changes and p values)

```python
#set up imports, define some functions
import pandas as pd
import numpy as np
import math

def extract_gene_id(x):
    info_parts = x.split('|')
    return info_parts[0]

def extract_gene_name(x):
    info_parts = x.split('|')
    return info_parts[1]

def log_trans(x):
    try:
        return math.log(x, 10)
    except:
        return float('NaN')
```

```python
#import salmon TPMs, and break up gene identifiers
gene_tpms_df = pd.read_csv('../20200408_combined_gene_TPMs.tsv', sep='\t')

gene_tpms_df['gene_id'] = gene_tpms_df['gene'].apply(extract_gene_id)
gene_tpms_df['gene_name'] = gene_tpms_df['gene'].apply(extract_gene_name)
```

```python
average_gene_tpms_df = gene_tpms_df[['gene', 'gene_id', 'gene_name']]
for genotype in ['WT', 'EDF1_ko']:
    for treatment in ['UN120', 'EM30', 'EM120']:
        av_sample = '{genotype}_{treatment}'.format(**locals())
        average_gene_tpms_df[av_sample] = sum([gene_tpms_df['{av_sample}_r{rep}'.format(**locals())] for rep in range(1,4)])/3.0
average_gene_tpms_df.to_csv('20200409_average_gene_TPMs.tsv', sep='\t', index=False)
```

```python
#plot fold changes
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
from adjustText import adjust_text
import math

plt.rcParams['pdf.fonttype'] = 42 #leaves most text as actual text in PDFs, not outlines
##plot CDF or mRNA RPKMs
black = (0,0,0)
gray = (0.5,0.5,0.5)
orange = (230/255.0,159/255.0,0)
skyBlue = (86/255.0,180/255.0,233/255.0)
bluishGreen = (0,158/255.0,115/255.0)
yellow = (240/255.0,228/255.0,66/255.0)
blue = (0,114/255.0,178/255.0)
vermillion = (213/255.0,94/255.0,0)
reddishPurple = (204/255.0,121/255.0,167/255.0)
colors = [black, orange, skyBlue, bluishGreen, vermillion, blue, reddishPurple, yellow]
rainbow = [black, vermillion, orange, bluishGreen, blue, reddishPurple, 'violet']

datasets = average_gene_tpms_df.columns.tolist()[3:]
num_plots_wide = len(datasets)
num_plots_high = num_plots_wide
fig = plt.figure(figsize=(2*num_plots_wide, 2*num_plots_high))
plot_index = 1

df = average_gene_tpms_df

for i in range(len(datasets)):
    for j in range(len(datasets)):
        plot = fig.add_subplot(num_plots_high, num_plots_wide, plot_index)
        if j < i:
            df.plot.scatter(datasets[j], datasets[i], color=black, s=1, alpha=1, ax=plot)
            plot.set_xlabel(datasets[j])
            plot.set_ylabel(datasets[i])
            plot.set_xscale('log', basex=2)
            plot.set_yscale('log', basey=2)
            plot.set_ylim(10**-3, 10**5)
            plot.set_xlim(10**-3, 10**5)
        elif j == i:
            #no point comparing same dataset
            plot.spines['left'].set_visible(False)
            plot.spines['bottom'].set_visible(False)
            plot.set_xticks([])
            plot.set_yticks([])
        else:
            #plot the correlations up top
            subset_df = pd.DataFrame()
            subset_df[datasets[i]] = df[datasets[i]].apply(log_trans)
            subset_df[datasets[j]] = df[datasets[j]].apply(log_trans)
            subset_df=subset_df.dropna()
            pr, pp = stats.pearsonr(subset_df[datasets[j]], subset_df[datasets[i]])
            sr, sp = stats.spearmanr(subset_df[datasets[j]], subset_df[datasets[i]])
            plot.set_xlabel(datasets[j])
            plot.set_ylabel(datasets[i])
            plot.text(-4.5, -2, 'r=%.3f\nrho=%.3f\n' % (pr, sr), fontsize=16)
            plot.set_ylim(-5, 5)
            plot.set_xlim(-5, 5)
            plot.set_xticks([])
            plot.set_yticks([])
        plot.spines['right'].set_visible(False)
        plot.spines['top'].set_visible(False)
        plot.set_aspect('equal')
        plot_index+=1

#lg=plt.legend(loc=4,prop={'size':12}, labelspacing=0.2)
#lg.draw_frame(False)
plt.tight_layout()
plt.savefig('20200409_average_gene_TPM_corr.pdf', transparent='True', format='pdf')
plt.savefig('20200409_average_gene_TPM_corr.png', transparent='True', format='png')
```

```python
deseq_comparisons = [EDF1ko_vs_WT_inUN, EM30_vs_UN_inWT, EM120_vs_UN_inWT, EM30_vs_UN_inEDF1ko, EM120_vs_UN_inEDF1ko, EM30_vs_UN_EDF1ko_interaction, EM120_vs_UN_EDF1ko_interaction]
comparison_names = ['UN_EDF1vWT', 'WT_EM30vUN', 'WT_EM120vUN', 'EDF1_EM30vUN', 'EDF1_EM120vUN', 'EDF1_effect_EM30', 'EDF1_effect_EM120']
sig_change_sets = {}
for index in range(len(deseq_comparisons)):
    df = deseq_comparisons[index]
    up = df[(df['padj']<0.01) & (df['log2FoldChange']>0.5)].dropna().sort_values('padj')['gene_name']
    df = deseq_comparisons[index]
    down = df[(df['padj']<0.01) & (df['log2FoldChange']<0.5)].dropna().sort_values('padj')['gene_name']
    bground = df.dropna().sort_values('padj')['gene_name']
    
    sig_change_sets['%s_up' % (comparison_names[index])] = set(up.tolist())
    sig_change_sets['%s_down' % (comparison_names[index])] = set(down.tolist())
    
    up.to_csv('GO_analysis/%s_GO_up.txt' %(comparison_names[index]), sep='\t', index=False, header=False)
    down.to_csv('GO_analysis/%s_GO_down.txt' %(comparison_names[index]), sep='\t', index=False, header=False)
    bground.to_csv('GO_analysis/%s_GO_background.txt' %(comparison_names[index]), sep='\t', index=False, header=False)

jun_targets = pd.read_csv('t0002-10.1080_15384047.2016.1156264.csv', sep=',')
cjun_cisplatin_targets = pd.read_csv('cjun-cisplatin-targets.csv', header=None)
cjun_colin_targets = pd.read_csv('cjun-targets-colin.csv', sep =',')

sig_change_sets['jun_targets']=set(jun_targets['Gene'])
sig_change_sets['cisplatin_jun_targets']=set(cjun_cisplatin_targets[0])
sig_change_sets['cjun_targets_colin']=set(cjun_colin_targets['Gene'])
```
